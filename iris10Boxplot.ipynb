```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')

print(df.head())

print(df.info())

print(df.dtypes)

plt.hist(df['sepal_length'], bins=20)

plt.title("Sepal Length Distribution")

plt.xlabel("Sepal Length")

plt.ylabel("Frequency")

plt.show()

plt.hist(df['sepal_width'], bins=20)

plt.title("Sepal Width Distribution")

plt.xlabel("Sepal Width")

plt.ylabel("Frequency")

plt.show()

plt.hist(df['petal_length'], bins=20)

plt.title("Petal Length Distribution")

plt.xlabel("Petal Length")

plt.ylabel("Frequency")

plt.show()

plt.hist(df['petal_width'], bins=20)

plt.title("Petal Width Distribution")

plt.xlabel("Petal Width")

plt.ylabel("Frequency")

plt.show()

sns.boxplot(y=df['sepal_length'])

plt.title("Sepal Length Boxplot")

plt.show()

sns.boxplot(y=df['sepal_width'])

plt.title("Sepal Width Boxplot")

plt.show()

sns.boxplot(y=df['petal_length'])

plt.title("Petal Length Boxplot")

plt.show()

sns.boxplot(y=df['petal_width'])

plt.title("Petal Width Boxplot")

plt.show()
```
